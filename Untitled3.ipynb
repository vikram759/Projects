

*   Bivariate analysis
*   categorical ->numerical and numerical->numerical
*   Use case - Time series data





```python
# plotting a simple function

price =[]
year=[2015,2016,2017,2018,2019,2020,2021]
price=[48000,54000,57000,49000,47000,45000,4500000]

# here 4500000 is overfitting

plt.plot(year,price)
```

```python
batsman=pd.read_csv('/content/sharma-kohli.csv')
batsman

plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title("Rohit Sharma vs Virat Kohli")
plt.xlabel("Year")
plt.ylabel("Runs")
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black')
# can use hex code too

plt.title("Rohit Sharma vs Virat Kohli")
plt.xlabel("Year")
plt.ylabel("Runs")
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='solid',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dashdot',linewidth=2)
# can use hex code too

plt.title("Rohit Sharma vs Virat Kohli")
plt.xlabel("Year")
plt.ylabel("Runs")
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='solid',linewidth=3,marker='D')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dashdot',linewidth=2,marker='')
# can use hex code too

plt.title("Rohit Sharma vs Virat Kohli")
plt.xlabel("Year")
plt.ylabel("Runs")
```

```python
#legend-> which line refers to whose

plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='solid',linewidth=3,marker='D',label='virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dashdot',linewidth=2,marker='D',label='rohit')
# can use hex code too

plt.title("Rohit Sharma vs Virat Kohli")
plt.xlabel("Year")
plt.ylabel("Runs")

plt.legend(loc='best')  #to show the taable bro
plt.show()
```

```python
price =[]
year=[2015,2016,2017,2018,2019,2020,2021]
price=[48000,54000,57000,49000,47000,45000,4500000]

# here 4500000 is overfitting

plt.plot(year,price)
plt.ylim(0,75000)
plt.xlim(2017,2019)
```

# **Bar Graph**

* Bvariate
* num vs categorical
* aggregate analysis of groups



```python
children=[10,20,30,13,67]
colors=['red','green','bllue','yellow','pink']

plt.bar(colors,children,color='black')
```

```python
plt.barh(colors,children,color='black') #horizontal bar graph
```

```python
# colors and labels

df=pd.read_csv('/content/batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'])

# dont benefit from this , bar graph overlap
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2)

plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
np.arange(df.shape[0])
```

```python
children=[10,20,30,13,67]
colors=['red re dre dre dre d re','green green green green green green green green','bllue','yellow','pink']

plt.bar(colors,children,color='black')

plt.xticks(rotation=90)
plt.show()
```

```python
# stacked bar chart

plt.bar(df['batsman'],df['2015'],label='Year 2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='Year 2016')
plt.bar(df['batsman'],df['2017'],bottom=df['2016']+df['2015'],label='Year 2017')

plt.legend()
```
